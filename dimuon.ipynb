```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.genfromtxt("dimuon_2-5GeV.csv", delimiter=',', dtype=None, names=True)
```

```python
m = data["M"]
plt.hist(m, bins=70, histtype='stepfilled', alpha=0.5);

x = np.linspace(2, 5, 1000)
y = 650*np.exp(-x)
plt.plot(x, y)
```

```python
s = [] # same charge
d = [] # different charge
for i, q1 in enumerate(data['Q1']):
    if q1 == data['Q2'][i]:
        s.append(data['M'][i])
    else:
        d.append(data['M'][i])
```

```python
plt.hist(s, bins=50, histtype='stepfilled', alpha=0.5);
plt.title("Same charge")
plt.show()
```

```python
plt.hist(d, bins=50, histtype='stepfilled', alpha=0.5);
plt.title("Different charge")
plt.show()
```

```python
eta1 = data['eta1']
pt1 = data['pt1']
eta2 = data['eta2']
pt2 = data['pt2']
```

```python
plt.plot(pt1, eta1, 'r.')
plt.title("First Muon")
plt.show()
```

```python
plt.plot(pt2, eta2, 'b.')
plt.title("Second Muon")
plt.show()
```
