```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.random(10000)
y = np.random.random(10000)
bins = 100
plt.hist(x, bins=bins, normed=1, color='b', histtype='stepfilled', alpha=0.5); 
plt.hist(y, bins=bins, normed=1, color='g', histtype='stepfilled', alpha=0.5);
```

```python
z = x*y
plt.hist(z, bins=bins, normed=1, color='r', histtype='stepfilled', alpha=0.5);

xp = np.linspace(0.01, 1, 1000)
zfit = -np.log(xp)
plt.plot(xp, zfit, 'b')
```

```python
z = x + y
plt.hist(z, bins=bins, normed=1, color='r', histtype='stepfilled', alpha=0.5);
```

```python
z = x / y
plt.hist(z, bins=20, normed=1, color='r', histtype='stepfilled', alpha=0.5);
```
