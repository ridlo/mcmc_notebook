## Probability Density Function (PDF)

+ Describe the likelihood of a variable to take on a given value.
+ Many PDF appears in nature, e.g. Poisson, Binomial, Normal, etc.

Example: Normal (a.k.a Gaussian) distribution

$f(x \vert \mu, \sigma) = \frac{1}{\sqrt{2 \pi} \sigma} \exp{-\frac{(x-\mu)^2}{2 \sigma^2}}, \quad x \in (-\infty, \infty)$

Notation for Normal distribution $= \mathcal{N}(\mu, \sigma^2)$

### Normalized PDF

$f(x) = \frac{1}{Z} \cdot f^{\star}(x)$

with

$Z = \int_{-\infty}^{+\infty} f^{\star}(x) dx$

Thus

$\int_{-\infty}^{+\infty} f(x) dx = 1$

Example of normalized Gaussian PDF:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline

x = np.linspace(-10,10,100)
plt.plot(x, norm.pdf(x, 0.0, 1.0), label=r'$\mu = 0, \sigma = 1.0$')
plt.plot(x, norm.pdf(x, 0.0, 2.0), label=r'$\mu = 0, \sigma = 2.0$')
plt.plot(x, norm.pdf(x, 0.0, 3.0), label=r'$\mu = 0, \sigma = 3.0$')
plt.plot(x, norm.pdf(x, 2.0, 1.5), label=r'$\mu = 2, \sigma = 1.5$')
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$f(x | \mu, \sigma)$', fontsize=18)
plt.legend(frameon=False, loc=0)
```

### Cumulative distribution function (CDF)

$F(x) = \int_{-\infty}^{x} f(x) dx$

Example of CDF of Normal distribution:

```python
x = np.linspace(-10, 10, 1000)

plt.plot(x, norm.cdf(x, 0.0, 1.0), label=r'$\mu = 0, \sigma = 1.0$')
plt.plot(x, norm.cdf(x, 0.0, 3.0), label=r'$\mu = 0, \sigma = 3.0$')
plt.plot(x, norm.cdf(x, 0.0, 5.0), label=r'$\mu = 0, \sigma = 5.0$')
plt.plot(x, norm.cdf(x, 2.0, 3.0), label=r'$\mu = 2, \sigma = 3.0$')
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$F(x | \mu, \sigma)$', fontsize=18)
plt.legend(frameon=False, loc=0)
```

### Inverse of CDF (quantile function)

$F^{-1}(p), \quad p \in (0, 1) $

Example of quantile function of Normal distribution (*probit function*):


```python
p = np.linspace(0, 1, 1000)

plt.plot(p, norm.ppf(p, 0.0, 1.0), label=r'$\mu = 0, \sigma = 1.0$')
plt.plot(p, norm.ppf(p, 0.0, 2.0), label=r'$\mu = 0, \sigma = 2.0$')
plt.plot(p, norm.ppf(p, 0.0, 3.0), label=r'$\mu = 0, \sigma = 3.0$')
plt.plot(p, norm.ppf(p, 2.0, 1.5), label=r'$\mu = 2, \sigma = 1.5$')
plt.xlabel(r'$p$', fontsize=18)
plt.ylabel(r'$F^{-1}(p|\mu, \sigma)$', fontsize=18)
plt.legend(frameon=False, loc=0)
```
