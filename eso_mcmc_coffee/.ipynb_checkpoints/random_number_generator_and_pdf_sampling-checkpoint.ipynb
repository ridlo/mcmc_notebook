## What is random number? 

Type of (uniform) random number:
1. **(nearly) true random number**
    - e.g. https://www.random.org/ ->  generate random number from atmospheric noise
    
2. **pseudo random number**
    - a sequence of numbers whose properties **approximate the properties of sequences of random numbers**.
    - is not truly random, because it is completely determined by an initial value, called the *seed* (which may include truly random values).
    - are important in practice for their **speed** in number generation and their **reproducibility**.
    
3. **quasi random (low-discrepancy sequence)**

---

+ each has its own "algorithm" to generate.
+ which one do you need? (depend on your case)

## (Pseudo) Random Number Generator

For basic purpose (**not for cyptography**), modern computer can generate **uniform random** (pseudo-random) with certain algorithm. <a href="https://en.wikipedia.org/wiki/Mersenne_Twister">Mersenne Twister</a> is a common algorithm.



#### Example of *uniform random number* generator: <a href="https://en.wikipedia.org/wiki/Linear_congruential_generator">Linear Congruental Generator</a> 

$$x_{n+1} = (a \cdot x_n + c ) \mod m$$

- easy to implement
    - obsolete but still used
    - poor choice of variable ($a, c, m$) can be dangerous, see <a href="https://en.wikipedia.org/wiki/RANDU">IBM's RANDU</a> case


```python
def makerandom(N, seed=1, modulus=10, a=2, c=3):
    res = []
    for i in range(N):
        seed = (a * seed + c) % modulus
        res.append(seed)
        
    return res
```

```python
makerandom(15)
```

For example `glibc` library use values: $a = 1103515245, c = 12345, m = 2^{31}$. It will repeat the sequence after $2^{31}$

What you need to remember:
- use a same *seed* if you want to reproduce the same result.
- be aware that the sequence will be back to the starting point again (after some large N generation)

## Tips: generating random number in polar/spherical coordinate
---

In many cases in astronomy, you may want to generate random number using *Polar* $(r, \theta)$ or *Spherical* coordinate $(r, \theta, \phi$). If you are not careful you will get the wrong distribution!

---
For example, the first thing you will likely do/think when you want to generate random variable in polar coordinate is simply by generating random number in the range of desired $r$ and $\theta$.

$u_1, u_2 =$ uniform random number [0, 1]

$\theta = 2 \pi u_1$ to get $\theta$ from $0$ to $2\pi$ 

$r = R u_2$ to get random number $r$ from $0$ to radius $R$

it is wrong!

```python
R = 5
N = 10000

theta = 2 * np.pi * np.random.random(N)
r = R * np.random.random(N)

x = r*np.cos(theta)
y = r*np.sin(theta)

plt.plot(x, y, 'k,')
plt.axis([-1.1*R, 1.1*R, -1.1*R, 1.1*R])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

The correct way:
$$\theta = 2 \pi u_1$$
$$r = R \sqrt{u_2}$$
    
Why? -> $dA = r d\theta dr$ (see inverse-transform method later)

```python
R = 5
N = 10000

theta = 2*np.pi*np.random.random(N)
r = R * np.sqrt(np.random.random(N))

x = r*np.cos(theta)
y = r*np.sin(theta)

plt.plot(x, y, 'k,')
plt.axis([-1.1*R, 1.1*R, -1.1*R, 1.1*R])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

For annulus shape

$$\theta = 2 \pi u_1$$

$$r = \sqrt{R_{min}^{2} + (R_{max}^{2} - R_{min}^{2})u_2}$$

```python
Rmin = 2
Rmax = 5
N = 10000

a = Rmin*Rmin
b = Rmax*Rmax - a

theta = 2*np.pi*np.random.random(N)
r = np.sqrt(a + b * np.random.random(N))

x = r*np.cos(theta)
y = r*np.sin(theta)

plt.plot(x, y, 'k,')
plt.axis([-1.1*R, 1.1*R, -1.1*R, 1.1*R])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

## Non Uniform Random Number Generator
---

In nature there are many "distributions", how can we generate random number with a certain "shape"?

> generate 'other distribution' from uniform distribution

**Methods :**
- Inverse transform sampling
- Acceptance-rejection sampling
- Markov Chain Monte Carlo (MCMC) sampling
- Special method/transformation --> for certain distribution e.g. Box-Muller transform for Gaussian distribution, ...

#### Example: *Gaussian random number* generator (<a href="https://en.wikipedia.org/wiki/Box%E2%80%93Muller_transform">Box-Muller Transform</a>)

A method to generate Gaussian random number from uniform random number


> Gaussian probability density function:

> $f(x \vert \mu, \sigma) = \frac{1}{\sqrt{2 \pi} \sigma} \exp{-\frac{(x-\mu)^2}{2 \sigma^2}}$, $x \in (-\infty, \infty)$


We can generate gaussian random number ($z_1$ or $z_2$) with $\mu = 0$ and $\sigma = 1$ from uniform random number $u_1$ and $u_2$

**Basic form:**

$u_1, u_2 =$ uniform random number [0, 1]

$z_1 = \sqrt{-2 \ln{u_1}} \cdot \cos{2 \pi u_2}$

$z_2 = \sqrt{-2 \ln{u_1}} \cdot \sin{2 \pi u_2}$

**Polar form:**

The polar form of the Box-Muller transformation is both faster and more robust numerically. It is faster because it does the equivalent of the sine and cosine geometrically *without a call to the trigonometric function* library. Included in *Numerical Recipes*.

with $u_1, u_2 = $ uniform random number [-1, 1] 

and $s = u_1^2 + u_2^2$

we can change the transformation into,

$z_1 = u_1 \sqrt{\frac{-2 \ln{s}}{s}}$

$z_2 = u_2 \sqrt{\frac{-2 \ln{s}}{s}}$

---

To generate random number in other $\mu$ and $\sigma$ we can do

$z = \mu + \sigma \cdot z_0$

where $z_0$ is the result from Box-Muller

---

> easy way in python, e.g. <pre>x = np.random.normal(size=10000)</pre>

```python
def random_normal():
    """Box-Muller transform: polar form
    mu = 0 and sigma = 1
    """
    s = 1.0
    while ( s >= 1.0 ):
        u1 = 2.0 * np.random.random() - 1.0; # -1 < u1 < 1
        u2 = 2.0 * np.random.random() - 1.0;
        s = u1*u1 + u2*u2

    w = np.sqrt(-2.0 * np.log(s) / s)
    z1 = u1 * w 
    # z2 = u2 * w
    return z1
```

```python
# generate
N = 10000
z = [random_normal() for i in range(N)]

# plot
from scipy.stats import norm
xp = np.linspace(-5,5,100)

plt.hist(z, bins=50, normed=True, alpha=0.5, linewidth=0, color='red');
plt.plot(xp, norm.pdf(xp))
plt.show()
```

```python
# other mu and sigma
mu = 3.0
sigma = 4.0
z1 = mu + sigma*np.array(z)

plt.hist(z, bins=50, normed=True, alpha=0.5, linewidth=0, color='red')
plt.hist(z1, bins=50, normed=True, alpha=0.5, linewidth=0, color='blue');
```

## PDF Sampling
---

In many cases, we want to sample a *probability density function* (PDF) with a certain 'shape'. Only a few well define PDF like Gaussian has a 'generator function'. Before you try the following methods, please check if your distribution already in the library, such as <a href="https://docs.scipy.org/doc/scipy/reference/stats.html">`scipy.stats`</a>

### Inverse transform sampling
---

We can sample a PDF by inversing it -> *if we can* inverse the *cumulative density function* (CDF). In many cases it is impossible/hard to get the inverse-CDF (ICDF / quantile function / percent-point function / PPF)

*Steps :* 
1. Calculate/find the CDF $F(x) = \int_{-\infty}^{x} f(x) dx$
2. Calculate inverse-CDF $F^{-1}(p), p \in (0, 1)$
3. Run:
    + Generate uniform random variable $u \in [0, 1]$
    + Calculate $x = F^{-1}(u)$

---

#### Example: Plummer distribution

According to <a href='https://en.wikipedia.org/wiki/Plummer_model'>Wikipedia</a>, Plummer 3-dimensional density profile is given by

$$ \rho(r) = \left( \frac{3M}{4\pi a^3} \right) \left( 1 + \left(\frac{r}{a}\right)^2 \right)^{-5/2} $$

> $r = $ radial distance from center

> $a = $ Plummer radius

---

By taking total mass $M = 1$, the mass enclosed within radius $r$ is given by

$$ m(r) = \int_{0}^{r} 4 \pi r^2 \rho(r) dr = \int_{0}^{r} \frac{3 r^2}{a^3} \left[ 1 + \left(\frac{r}{a}\right)^2 \right]^{-5/2} dr = \frac{(r/a)^{3}}{\left[ 1 + (r/a)^{2} \right]^{3/2}}$$

with $M = 1$ that equation become our CDF!

$$F(r | a) = m(r) = \frac{(r/a)^{3}}{\left[ 1 + (r/a)^{2} \right]^{3/2}}$$

---
Plummer PDF by definition then can be written as (equation inside the integral above)

$$ f(r | a) = \frac{3 r^2}{a^3}\left[1 + \left(\frac{r}{a}\right)^2\right]^{-5/2} $$

We can find the inverse-CDF 

Take $x = r/a$
$$
\begin{eqnarray}
m &=& \frac{x^3}{(1 + x^2)^{3/2}}\\
x^3&=& m (1 + x^2)^{3/2} \\
x^2 &=& m^{2/3} (1 + x^2) \\
\frac{1 + x^2}{x^2} &=& m^{-2/3}\\
\frac{1}{x^2} + 1 &=& m^{-2/3}\\
x = r/a &=& (m^{-2/3} - 1)^{-1/2}
\end{eqnarray}
$$

The inverse CDF of Plummer distribution:
$$F^{-1}(m | a) = r(m) = a\left( m^{-2/3} - 1\right)^{-1/2}$$

```python
def plummer(r, a=1.0, arg='ppf'):
    if arg == 'density': 
        return (3/(4*np.pi*a*a*a)) * np.power((1 + r*r/(a*a)), -5./2)
    elif arg == 'pdf': 
        return (3*r*r/(a*a*a)) * np.power((1 + r*r/(a*a)), -5./2.)
    elif arg == 'cdf': 
        x = r/a
        return x*x*x * np.power((1 + x*x), -3./2.)
    elif arg == 'ppf': 
        return a / np.sqrt(np.power(r, -2./3.) - 1.0) # r = m
    else:
        print('Error arg')
        return None

def plot_plummer(a=1, arg='pdf', rmin=0.0, rmax=10.0, N=1000, xlabel=r'$r$', ylabel=r'$f(r|a)$'):
    r = np.linspace(rmin, rmax, N)
    f = plummer(r, a, arg)

    plt.plot(r, f)
    plt.xlim([0, rmax+0.01])
    plt.title(arg); plt.xlabel(xlabel); plt.ylabel(ylabel);
    plt.show()
```

```python
a = 1
r0 = np.linspace(0.0, 10, 100)
r_ = np.linspace(0.00001, 0.99, 100) # for ppf
r = [r0, r0, r0, r_]
args = ['density', 'pdf', 'cdf', 'ppf']
xlbl = [r'$r$', r'$r$', r'$r$', r'$m$']
ylbl = [r'$\rho(r|a)$', r'$f(r|a)$', r'$F(r|a)$', r'$F^{-1}(m|a)$']
co   = ['r', 'g', 'b', 'k']
matplotlib.rcParams.update({'font.size': 14})

xSize = 15
ySize = xSize/GOLDEN_RATIO
nRow, nCol = 2, 2

fig = plt.figure(figsize=(xSize, ySize))
gs = gridspec.GridSpec(nRow, nCol)

k = 0
for i in range(nRow):
    for j in range(nCol):
        ax = plt.subplot(gs[i, j])
        ax.plot(r[k], plummer(r[k], a, args[k]), co[k])
        ax.set_xlabel(xlbl[k]); ax.set_ylabel(ylbl[k]); ax.legend([args[k]])
        k += 1
```

```python
a = 1.0
npoints = 20000
rmin, rmax = 0.0, 10.0
nbins=100

# inverse-transform sampling
rsample = plummer(np.random.random(npoints), a, 'ppf')

r = np.linspace(rmin, rmax, nbins)
f = plummer(r, a, 'pdf')

# plot
plt.hist(rsample, bins=nbins, normed=True, range=[rmin, rmax], color='red', histtype='stepfilled', alpha=0.5);
plt.plot(r, f)
plt.show()
```

### Acceptance-rejection sampling
---
Acceptance-rejection method assume we have a simpler distribution $q^*(x)$ which we can easily draw samples. Our proposal distribution $q^*(x)$ must cover targeted distribution/PDF $p^*(x)$ in our interested region $(x_{min}, x_{max})$.

$$q^*(x) > p^*(x), \quad \text{for all } x \text{ or interested region } (x_{min}, x_{max})$$

$Steps :$
- generate random variable $x$ in our interested region $[x_{min}, x_{max}]$
- evaluate $p^*(x)$ 
- generate random variable $y$ in the range $[0, q^*(x)]$
- if $y < p^*(x)$, then accept $x$, otherwise reject $x$.


---
#### Example:

Let's try to generate PDF $p^*(x)$ using *several proposal distributions* $q^*(x)$, 

$$p^*(x) = C \cdot e^{0.4(x-0.4)^2 - 0.08x^4}$$

> C = a constant so that integral of this function is 1 --> pdf

> with a numerical method (e.g. <code>scipy.integrate.quad</code>) we can find the value of $C \approx 1/7.852178178866215 = 0.12735319770142955$

you can try to integrate this function analytically if you want ;)

---

```python
def pdftarget(x):
    return 0.12735319770142955 * np.exp(0.4 * (x - 0.4)**2 - 0.08 * x**4)
```

```python
# plot the pdf
xmin, xmax = -5, 5
N = 100
x = np.linspace(xmin, xmax, N)
y = pdftarget(x)

plt.plot(x, y)
plt.show()
```

**1. Uniform dist as proposal distribution**

```python
def sampling_from_uniform(func, nsamples=1000, xmin=-5, xmax=5, c=1):
    # uniform proposal distribution
    xsample = xmin + (xmax - xmin)*np.random.random(nsamples) 
    target_dist = func(xsample)
    ysample = c*np.random.random(nsamples) # 0 < y < c
    
    # accept if ysample < y(x)
    accept = [ysample <= target_dist]
    
    return xsample, ysample, accept
```

```python
# montecarlo to estimate the highest point
x = np.linspace(xmin, xmax, 1000)
p = pdftarget(x)
yhighest = np.amax(p)
print(yhighest) # it is not the real ymax

# choose a higher value in y. You must be sure that this value is higher than ymax of the pdf
yc = 1.01*yhighest
print(yc)
```

```python
xsample, ysample, accept = sampling_from_uniform(pdftarget, nsamples=1000000, xmin=-5, xmax=5, c=yc)
xaccept = xsample[accept]

# plot
plt.hist(xaccept, bins=100, normed=1, histtype='stepfilled', color='red', alpha=0.5);
plt.plot(x, p, 'k')
plt.axhline(y=yc, color='g')
```

**2. Gaussian dist as proposal distribution**

Find a gaussian that 'cover' our distribution function 

```python
x = np.linspace(xmin, xmax, 1000)
target_dist = pdftarget(x)

mu = -1
sigma = 2
propos_dist = norm.pdf(x, mu, sigma)

xhighest_idx = np.argmax(target_dist)
yhighest = np.amax(target_dist)
c = 1.05*yhighest/propos_dist[xhighest_idx]

print(mu, sigma, c)
    
plt.plot(x, target_dist)
plt.plot(x, c*propos_dist)
plt.xlim([xmin, xmax])
plt.show()
```

```python
def sampling_from_gaussian(func, mu=0, sigma=1, c=2, nsamples=1000, xmin=-5, xmax=5):
    """Please find a good proposal dist before running this function"""
    
    # start sampling
    xsample = np.random.normal(mu, sigma, nsamples) 
    target_dist = func(xsample)
    ypropos = c * norm.pdf(xsample, mu, sigma)
    ysample = ypropos * np.random.random(nsamples) # uniform in y
    # compare
    accept = [ysample <= target_dist]
    
    return xsample, ysample, accept
```

```python
xsample, ysample, accept = sampling_from_gaussian(pdftarget, mu=-1.0, sigma=2.0, c=c, nsamples=1000000)

plt.hist(xsample[accept], bins=50, normed=1, histtype='stepfilled', linewidth=0, color='red', alpha=0.5);
plt.plot(x, target_dist)
plt.plot(x, c*propos_dist)
plt.xlim([xmin, xmax])
plt.show()
```

### Markov Chain Monte Carlo (MCMC) sampling
---

We use a proposal distribution $Q(x' | x^{(t)})$ to generate a proposed next state $x'$ based on current state $x^{(t)}$.

#### Example: <a href="https://en.wikipedia.org/wiki/Metropolis%E2%80%93Hastings_algorithm">Metropolis-Hastings algorithm</a>

*Steps:*

1. Initiate state $x^{(t=0)}$
2. Generate a proposed state from $Q(x' | x^{(t)})$
3. Compute $a = \frac{f^{*}(x')}{f^{*}(x^{(t)})} \frac{Q(x^{(t)} | x')}{Q(x' | x^{(t)})}$
4. If $a \ge 1$ accept proposed state $x'$
5. Otherwise, generate $u$ from a uniform distribution in interval $[0, 1]$ and accept the proposed state if $a \ge u$.
6. If the state is accepted, we set $x^{(t+1)} = x'$
7. If the state is rejected, stick to the previous state and set $x^{(t+1)} = x^{(t)}$

> Note that if $Q(x' | x^{(t)})$ is symmetric, the ratio of $\frac{Q(x^{(t)} | x')}{Q(x' | x^{(t)})}$ is always 1.


```python
def sampling_mcmc_mh(func, xt, stepsize, nsamp):
    """Metropolis-Hastings algorithm
    xt = mu
    stepsize = sigma
    """
    
    samples = np.empty(nsamp)
    accept = np.empty(nsamp)
    
    # start MCMC iteration
    for i in range(nsamp):
        xprime = xt + stepsize*np.random.normal() # gaussian proposal distribution to propose a new step
        a = func(xprime)/func(xt) # symmetric -> gaussian
        if a >= 1.0:
            xt = xprime
            accept[i] = 1
        else:
            u = np.random.random()
            if a >= u:
                xt = xprime
                accept[i] = 1
            else:
                accept[i] = 0 # reject xprime, xt = xt
        
        samples[i] = xt
        
    return samples, accept
```

```python
mu = 0
sigma = 2
nsamp = 100000

samples, accept = sampling_mcmc_mh(pdftarget, mu, sigma, nsamp)

# plot
plt.hist(samples, bins=50, normed=1, histtype='stepfilled', linewidth=0, color='red', alpha=0.5);
plt.plot(x, target_dist)
plt.xlim([xmin, xmax])
plt.show()
```
